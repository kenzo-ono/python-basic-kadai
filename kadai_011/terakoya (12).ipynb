```python
array = ["水","金","地","火","木","土","天","海","冥"]

# forループ
for value in array:
    print(value)

# whileループ
i = 0
while i < len(array):
    print(array[i])
    i = i + 1
```

```python
# 判定する数値を変数varに代入
var = 15  # ←好きな正の整数に変更してOK

# 条件分岐（if / elif / else）
if var % 3 == 0 and var % 5 == 0:
    print("FizzBuzz")
elif var % 3 == 0:
    print("Fizz")
elif var % 5 == 0:
    print("Buzz")
else:
    print(var)
```
